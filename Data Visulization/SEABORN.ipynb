```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
tips.head()
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    # size="size",
    # style="smoker"
)
```

```python

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
    hue="smoker",
)
```

```python
# scatter plots
tips=sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
# LINE PLOTS

flights=sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
# BAR PLOTS
tips=sns.load_dataset("tips")

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
# BOX PLOT

tips=sns.load_dataset("tips")

sns.boxplot(
    data=tips,
    x="day",
    y="tip",
    # hue="sex"
)
```

```python

penguins=sns.load_dataset("penguins")
penguins.head()

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=20
)
```

```python
# matrix plots

# heatmaps- it is a plot that use for rectenglar data as color encoded metrix

flights=sns.load_dataset("flights")
flights_pivot=flights.pivot(index="month",columns="year",values="passengers")
sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
plt.title("Passengers Heatmap example")
plt.tight_layout()
```

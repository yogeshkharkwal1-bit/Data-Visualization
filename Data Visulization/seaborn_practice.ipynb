```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")

fig,ax=plt.subplots()


sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
    errorbar=None
)
ax.set_title("avg bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("bills")

fig.tight_layout()

```
